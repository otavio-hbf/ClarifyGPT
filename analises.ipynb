```python
import json

# Open and read the JSONL file
with open(r'data\clarifygpt_mbpp\mbpp_askcq_three_shot_chatgpt_results.jsonl', 'r', encoding='utf-8') as file:
    total_tokens_list = []
    
    for line in file:
        data = json.loads(line)
        
        total_tokens = data[2]['usage']['total_tokens']
        total_tokens_list.append(total_tokens)


# Summary statistics
print(f"\nTotal entries: {len(total_tokens_list)}")
print(f"Sum of all tokens: {sum(total_tokens_list)}")
print(f"Average tokens: {sum(total_tokens_list) / len(total_tokens_list) if total_tokens_list else 0}")
print(f"Max tokens: {max(total_tokens_list) if total_tokens_list else 0}")
print(f"Min tokens: {min(total_tokens_list) if total_tokens_list else 0}")
```
